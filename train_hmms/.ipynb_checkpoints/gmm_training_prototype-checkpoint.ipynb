```python
state1 = State(PoissonDistribution(3))
state2 = State(PoissonDistribution(20))
model = HiddenMarkovModel(name='CTCF_Peaks')
```

```python
model.add_transition(model.start,state1,0.01)
model.add_transition(model.start,state2,0.99)
model.add_transition(state1, state1, 0.7)       # transition probability from state1 to state1
model.add_transition(state1, state2, 0.3)       # transition probability from state1 to state2
model.add_transition(state2, state1, 0.01)       # transition probability from state2 to state1
model.add_transition(state2, state2, 0.99)       # transition probability from state2 to state2
model.add_transition(state1, model.end, 0.1)    # transition probability from state1 to end
model.add_transition(state2, model.end, 0.1)    # transition probability from state2 to end
```

```python
model.bake()
```

```python
model.fit(list_of_sequences,max_iterations=100,stop_threshold=0.001)
```

```python
for sequence in list_of_sequences:
    state_sequence = model.predict(sequence)
    log_prob = model.log_probability(sequence)
#     state_names = [model.states[index].name for index in state_sequence]
    plt.plot([state for state in state_sequence])
    plt.plot(sequence)
    plt.show()
#     print("Most likely sequence of states:", state_sequence)
#     print("Log probability of the sequence:", log_prob)
#     print()
```
